# Univariate , bivariate , multivariate

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    "sepal_length" :[5 , 6 , 7 , 8 , 9],
    "sepal_width" :[3,2,1,4,5],
    "petal_length" :[2,3,4,5,6],
    "petal_width" : [2,4,2,1,1],
    "species" : ["versicolor" , "verginica" , "setosa" , "setosa" , "setosa"]
})
```

```python
df
```

```python
df.shape
```

```python
df.loc[df['species'] == 'setosa']
```

```python
df_versi = df.loc[df['species'] == 'versicolor']
df_virgin = df.loc[df['species'] == 'verginica']
```

```python
%matplotlib inline
```

```python
plt.plot(df['sepal_length'] , np.zeros_like(df['sepal_length']), 'o')
plt.plot(df_versi['sepal_length'] , np.zeros_like(df_versi['sepal_length']),'o')
plt.xlabel('Petal length')
plt.plot(df_virgin['sepal_length'] , np.zeros_like(df_virgin['sepal_length']),'o')
```

```python
sns.FacetGrid(df , hue = "species" ).map(plt.scatter , "sepal_length" , "sepal_width").add_legend()
```

```python
sns.pairplot(df , hue = "species")
```
