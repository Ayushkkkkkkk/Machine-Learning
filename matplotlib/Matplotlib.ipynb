```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x = ["Sunday" , "Monday" , "Tuesday" , "Wednesday" , "Thrusday" , "Friday" ,"Saturday"]
y = [50 , 51 , 53, 65, 45 , 90 , 100]
```

```python
%matplotlib inline
plt.xlabel("DAY")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x , y ,'--r*')
plt.grid()
```

```python
plt.plot(x, y , color = "red" ,marker = 'D' , linestyle = '-')
plt.grid()
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,55,56,57,58,69,70]
min_t = [32,23,43,42,23,12,23]
avg_t = [43,32,44,56,43,34,23]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days , max_t , label = "MAX")
plt.plot(days , min_t , label = "MIN")
plt.plot(days , avg_t , label = "AVG")
plt.legend(loc = "best" , shadow = True)
plt.grid()
```

```python
company = ["Google" , "Amazon" , "Facebook" , "Microsoft"]
revenue = [400 , 500 , 600 , 700]
profit = [100 , 200 , 300 , 400]
```

```python
plt.plot(company,revenue) 
```

```python
ypos = np.arange(len(company))
plt.xticks(ypos , company)
plt.ylabel("Revenue")
plt.title("US TECH STOCKS")
plt.bar(ypos, revenue , label = "Revenue")
plt.bar(ypos , profit , label = "profit")
plt.legend(loc = "best")
```

Histogram

```python
blood_sugar = [100 , 113, 114 , 120 , 90 , 80 , 89 , 98]

plt.hist(blood_sugar, bins = 3 , rwidth = 0.95 , color = "green")
```

```python
blood_sugar_man = [100 , 113, 114 , 120 , 90 , 80 , 89 , 98]
blood_sugar_woman = [150 , 190, 176 , 76 , 65 , 76 , 54 , 45]
```

```python
plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_man , blood_sugar_woman], bins = 3 , rwidth = 0.95 , color = ["green" , "orange"] , label = ["man" , "woman"])
plt.legend()
```

```python
expenses = [100 , 200 , 400 , 1000]
usage = ['internet' , "food" , "clothes" , "holidays"]
```

```python
plt.pie(expenses , labels = usage , radius = 1 , autopct ='%.2f%%' , shadow = True )
plt.show()
plt.savefig("piechart.png" , bbox_inches = "tight")
```
