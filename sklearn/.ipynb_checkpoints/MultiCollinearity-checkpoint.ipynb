```python
import pandas as pd
import statsmodels.api as sm
df_adv = pd.read_csv("Advertising.csv" , index_col = 0)
s = df_adv[['TV' , 'radio' , 'newspaper']]
y = df_adv['sales']
```

```python
X = sm.add_constant(s)
model = sm.OLS(y , s).fit()
```
